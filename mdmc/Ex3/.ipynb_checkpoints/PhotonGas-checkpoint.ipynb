# The Photon Gas

This exercise gives you an example of a Metropolis Monte Carlo algorithm, where you will apply this type of algorithm to calculate the state occupancy of a photon gas. The photon gas is a gas-like collection of photons, which has many of the same
properties of a conventional gas such as pressure, temperature and
entropy. The most common example of a photon gas in equilibrium is
black-body radiation. Black-body radiation is an electromagnetic field
constructed by a superposition of plane waves of different frequencies,
with the caveat that a mode may only be excited in units of $\hbar w$.
This fact leads to the concept of photons as quanta of the
electromagnetic field, with the state of the field being specified by
the occupancy $\left<n_j\right>$ of each of the modes or, in other
words, by enumerating the number of photons with each frequency.

## Ensemble Averages from the Metropolis Monte Carlo Algorithm

The ensemble average of the state occupancy $\left<n_j\right>$ of a
photon gas can be calculated analytically. Deriving the total energy of
an idealised photon gas from quantum mechanics we know that $U$ can be
written as the sum of the harmonic oscillation energies:

$$
\begin{aligned}
U= \sum_{j=1}^{N} n_j w_j \hbar = \sum_{j=1}^{N} n_j \epsilon_j,
\end{aligned}
$$ (total_energy_photongas)

where $w_j$  and $\epsilon_j$ are the oscilator frequency and energy of state $j$, $n_j$ is the occupancy of
state $j$ ($n_j \in 0,1,2,\cdots, \infty$), and $N$ is the total number of
states. In this exercise, you are
going to compute the ensemble average of the occupancy
$\left<n_j\right>$. 

For simplicity, we will assume that all the frequencies are the same, i.e. $\epsilon_j = \epsilon$ for every $j$ and hence the energy expression simplifies to

$$
\begin{aligned}
U= \epsilon \sum_{j=1}^{N} n_j .
\end{aligned}
$$ (total_energy_photongas) 

The scheme you will employ is as follows:

1.  Start with an arbitrary $n_j$.

2.  Decide to perform a trial move to randomly increase or decrease
    $n_j$ by 1.

3.  Accept the trial move with probability (i.e. perform a Metropolis Monte Carlo move):Â 

    $$
    \begin{aligned}
    P_{acc}(o \rightarrow n)= \min \left(1, e^{-\beta(U(n)-U(o))}\right),\end{aligned}
    $$
    
    where $U(n)$ and $U(o)$ are the energies of the new and old states
    respectively.

4.  Update averages regardless of acceptance or rejection.

5.  Iterate from 2).

**Note** that in the code we will measure the energy in units of $\hbar w$. This means that when you calculate the energy, you will only need to enumerate the number of photons with each frequency since the the Boltzmann's factor is given by $e^{-\beta(U(n)-U(o))}$.

```python
numberOfIterations = 1000
beta = 1.0


def calculateOccupancy(beta=1.0):

    trialnj = 1
    currentnj = 1
    njsum = 0
    numStatesVisited = 0
    estimatedOccupancy = 0

    """  MODIFICATION
    Metropolis algorithm implementation to calculate <n_j>
    Tasks:
    1) Loop from int i = 0 to numberOfiterations
    2) Call r.randint(0,1) to perform a trial move to randomly increase
        or decrease trialnj by 1.
        Note: randint(0,1) returns random integers from 0 to 1 (i.e. 0 OR 1)
               but you need to extract -1 OR 1
        Hint:   what happens to the extraction if you multiply r.randint(0,1)*A, where A is an integer?
                what happens to the extraction if you sum r.randint(0,1)+B, where B is an integer (positive or negative)?
                The above hints shall suggest how to obtain the desired numbers from the extraction of 0 or 1
    3) Test if trialnj < 0, if it is, force it to be 0
    4) Accept the trial move with probability defined in the Theory section
        Note: Accepting the trial move means updating current sample (currentnj)
        with the new move (trialnj);
    5) sum currentnj and increase numStatesVisited by 1
    6) compute estimatedOccupancy after numberOfiterations iterations
    
    END MODIFICATION
    """
    # INSERT YOUR CODE HERE

    # Loop from int i = 0 to numberOfiterations
    for i in range(numberOfIterations):

        # Step 2: Propose a trial move
        # r.randint(0,1) gives 0 or 1; transform to -1 or +1
        move = r.randint(0, 1) * 2 - 1  # Results in -1 (decrease) or +1 (increase)
        trialnj = currentnj + move

        # Step 3: Enforce non-negative occupancy
        if trialnj < 0:
            trialnj = 0

        # Accept move if probability is high

        # Step 4: Calculate energy difference and acceptance probability
        Uo = currentnj
        Un = trialnj
        deltaU = Un - Uo  # energy difference
        P_acc = min(1, np.exp(-beta * deltaU))  # Probablity of acceptance

        # Accept or reject the move
        if r.random() < P_acc:
            currentnj = trialnj

        # Step 5: Update running sum and counter
        njsum += currentnj
        numStatesVisited += 1

    # Step 6: Compute the estimated occupancy
    estimatedOccupancy = njsum / numStatesVisited

    return estimatedOccupancy


# perform a single calculation
estimatedOccupancy = calculateOccupancy(beta=beta)
```

```python
def BatchCalculateOccupancy(
    beta=1.0, numberOfIterations=1000, numChains=100, modified=False
):

    # Random initialization of states
    currentnj = np.ones(numChains)  # Array of initial states
    njsum = np.zeros(numChains)  # Total sum of all current nj's
    numStatesVisited = np.zeros(numChains)  # Number of states visited

    # Pre-generate random moves (+1 or -1) for all iterations
    moves = np.random.choice([-1, 1], size=(numChains, numberOfIterations))

    # Pre-generate random acceptance probabilities
    rand_probs = np.random.rand(numChains, numberOfIterations)

    # Loop through all iterations
    for i in range(numberOfIterations):
        # Compute trialnj for all chains simultaneously by adding the move
        trialnj = currentnj + moves[:, i]

        # Force trialnj < 0 to 0 (vectorized)
        trialnj = np.maximum(trialnj, 0)

        # Compute the energy difference (delta U) for all chains
        deltaU = trialnj - currentnj

        # Calculate acceptance probability for all chains
        P_acc = np.minimum(1, np.exp(-beta * deltaU))

        # Accept or reject the move (vectorized)
        accepted = rand_probs[:, i] < P_acc
        
 
        # Update sum and number of states visited
        if modified:

            ### CHANGE HERE (comment or uncomment) ###
            mask = accepted
            # mask = accepted & (trialnj != currentnj)
            #######
            
            currentnj[mask] = trialnj[mask]  # Update accepted chains
            njsum[mask] += currentnj[mask] # Update sum only on acceptance
            numStatesVisited[mask] += 1 # Count only accepted moves
        else:
            currentnj[accepted] = trialnj[accepted]  # Update accepted chains
            njsum += currentnj
            numStatesVisited += 1

    # Compute estimated occupancy (average occupancy) where numStatesVisited > 0 to avoid division by zero
    estimatedOccupancy = np.divide(
        njsum, numStatesVisited, out=np.zeros_like(njsum), where=numStatesVisited > 0
    )
    return estimatedOccupancy


numChains = 20  # Number of random initial states
np.random.seed(42)

betas = np.linspace(0.1, 2)
analytical_y = 1 / (np.exp(betas) - 1)

cmap = plt.get_cmap("tab10")
fig, ax = plt.subplots(1)
ax.set_title(f"Photon-Distribution Plot averaged with {numChains} chains")
ax.plot(betas, analytical_y, label="analytical")


for idx, numIterations in enumerate([10, 100, 1000]):
    for idx_b, b in enumerate(betas):
        estimatedOccupancy = BatchCalculateOccupancy(b, numIterations, numChains)
        ax.scatter(
            b,
            np.mean(estimatedOccupancy),
            marker="o",
            s=10,
            color=cmap(idx),
            label=f"{numIterations} Iterations" if idx_b == 0 else "",
        )

ax.set_xlabel("beta")
ax.set_ylabel("Occupancy")
ax.grid(True)
ax.legend()
plt.tight_layout()
plt.savefig('ex5.2.png')
plt.show()
```

**Answer**
Increasing the number of Monte Carlo iterations reduces statistical fluctuations, leading the estimated result to converge closer to the analytical value. This occurs because more samples average out random errors, in accordance with the Law of Large Numbers.

**Goal**

We want to compute the average occupancy number $\langle n \rangle$ of a state. In equilibrium, the true average is given by

$$
\langle n \rangle = \sum_n \mathcal{N}(n) \, n,
$$

where $\mathcal{N}(n)$ is the probability (the Boltzmann weight $\exp(-\beta U(n)) / Z$ ) of finding the system in state $n$.

**How We Should Average**

Because our chainâs states (whether they change or not) are sampled according to the correct equilibrium distribution, the proper way to estimate an average is to include every Monte Carlo step. That is, after $T$ steps, we compute

$$
\langle n \rangle_{\text{MC}} = \frac{1}{T} \sum_{t=1}^{T} n_t,
$$

where $n_t$ is the state at step $t$. In this way, the chain âremembersâ when it did not change state (a rejection), and this is very important.

**The âAccepted-Onlyâ Modification**

We now only update the running average when a trial move is accepted. This means that if a move is rejected, you do nothing, and you do not record the state at that step. Then your average is computed as

$$
\langle n \rangle_{\text{acc}} = \frac{1}{T_A} \sum_{t \in A} n_t,
$$

where $A$ is the set of steps at which the move was accepted and $T_A$ is the number of accepted moves. When you count only accepted moves, the effective probability $\mathcal{N}_{\text{eff}}(n)$ that you record state $n$ is not simply $\mathcal{N}(n)$; it is modified by the probability that a move from $n$ is accepted. Letâs denote the probability that a move is accepted when you are in state $n$ by $P_A(n)$.

Then the effective weight for state $n$ becomes

$$
\mathcal{N}_{\text{eff}}(n) \propto \mathcal{N}(n) P_A(n).
$$

Unless $P_A(n)$ is a constant (i.e., independent of $n$), the effective distribution $\mathcal{N}_{\text{eff}}(n)$ is not equal to the true equilibrium distribution $\mathcal{N}(n)$. Since upward moves (from $o$ to $n$) are often rejected, you end up recording many occurrences of $o$ only when a move out of $o$ is accepted. This biases your sampling: states with a lower acceptance probability (like $o$ in the upward move scenario) get under-represented in the âaccepted-onlyâ record.

Thus, your average computed from accepted moves is

$$
\langle n \rangle_{\text{acc}} = \frac{\sum_n \mathcal{N}(n) P_A(n) n}{\sum_n \mathcal{N}(n) P_A(n)},
$$

which is not the same as

$$
\langle n \rangle = \sum_n \mathcal{N}(n) n,
$$

unless $P_A(n)$ is constant for all $n$.

At which values of $\beta$ does the error you obtain when ignoring rejected moves become more pronounced? Why?

**Dependence on $\beta$**

Recall that in the Metropolis scheme, for a move that increases the energy (i.e., an upward move, which increases $n$), the acceptance probability is

$$
\text{acc}(o \to n) = \exp \{-\beta [U(n) - U(o)]\}.
$$

If we set $\hbar \omega = 1$ and use $U(n) = n$ (for simplicity), then a move that increases $n$ by 1 is accepted with probability

$$
\exp(-\beta).
$$

For a move that decreases the energy, the acceptance probability is 1.

**Effect at Different Values of $\beta$**

**At small $\beta$ (high temperature):**

$\exp(-\beta)$ is close to 1. So, both upward and downward moves are almost always accepted. In this case, $P_A(n) \approx 1$ for all $n$. The result is that the bias in accepted-only averaging is minimal.

**At large $\beta$ (low temperature):**

$\exp(-\beta)$ is very small. Thus, upward moves (which increase $n$) are very rarely accepted, while downward moves are always accepted.  
This makes $P_A(n)$ state-dependent: states where most moves are upward (or where upward moves are more likely) will have a lower $P_A(n)$ than states where downward moves dominate. The result is that the effective sampling $\mathcal{N}_{\text{eff}}(n) \propto \mathcal{N}(n) P_A(n)$ becomes significantly different from $\mathcal{N}(n)$. In particular, states with higher $n$ (which require many upward moves) will be under-represented in the accepted-move record.

Thus, the error (bias) caused by averaging only on accepted moves becomes more pronounced as $\beta$ increases (i.e., at lower temperatures).

## Appendix: Computing averages from the MCMC

Suppose we want to compute the average over the posterior of an observable $A(s)$
To do this we use the fact that (under the ergodicity assumption) the time averages over in the Markov chain converge to the ensemble averages. In formulas we have
$$
\lim_{T\to\infty} \frac{1}{T}\sum_{t=0}^T A(s^{(t)})=\langle A \rangle=\sum_{s} P(s|Y) A(s)
$$

In practicve things are more involved.

As you saw, the Monte Carlo needs some time to thermalize, before it starts sampling from the equilibrium measure. Configurations encountered in this initial *burn in* phase are not representative of the posterior, thus they should be discarded.
Let $T_{therm}$ be the time until the Markov chain has reached equilibrium, and $T$ the maximum time of the simulation.

Then we approximate
$$
\langle A \rangle \approx \frac{1}{T-T_{therm}}\sum_{t=T_{therm}}^T A(s^{(t)})
$$

```python
import numpy as np
import matplotlib.pyplot as plt

def analytical_n(beta_eps):
    return 1 / (np.exp(beta_eps) - 1)

def monte_carlo_n_original(beta_eps, steps=1000000, burn_in=10000):
    n = 0
    n_sum = 0
    accepts = 0
    for _ in range(burn_in):
        move = np.random.choice([-1, 1])
        n_new = n + move
        if n_new < 0:
            n_new = 0
        delta_U = n_new - n
        if delta_U <= 0:
            accept = True
        else:
            accept_prob = np.exp(-beta_eps * delta_U)
            accept = np.random.random() < accept_prob
        if accept:
            n = n_new
    for _ in range(steps):
        move = np.random.choice([-1, 1])
        n_new = n + move
        if n_new < 0:
            n_new = 0
        delta_U = n_new - n
        if delta_U <= 0:
            accept = True
        else:
            accept_prob = np.exp(-beta_eps * delta_U)
            accept = np.random.random() < accept_prob
        if accept:
            n = n_new
            accepts += 1
        n_sum += n
    # print(f"Original Î²Ïµ={beta_eps:.2f}, Acceptance Rate: {accepts/steps:.3f}")
    return n_sum / steps

def monte_carlo_n_modified(beta_eps, steps=1000000, burn_in=10000):
    n = 0
    n_sum = 0
    accepted_steps = 0
    for _ in range(burn_in):
        move = np.random.choice([-1, 1])
        n_new = n + move
        if n_new < 0:
            n_new = 0
        delta_U = n_new - n
        if delta_U <= 0:
            accept = True
        else:
            accept_prob = np.exp(-beta_eps * delta_U)
            accept = np.random.random() < accept_prob
        if accept:
            n = n_new
    for _ in range(steps):
        move = np.random.choice([-1, 1])
        n_new = n + move
        if n_new < 0:
            n_new = 0
        delta_U = n_new - n
        if delta_U <= 0:
            accept = True
        else:
            accept_prob = np.exp(-beta_eps * delta_U)
            accept = np.random.random() < accept_prob
        if accept and n_new != n:  # Only count if state changes
            n = n_new
            n_sum += n
            accepted_steps += 1
        elif accept:
            n = n_new  # Update state but don't count in average
    # print(f"Modified Î²Ïµ={beta_eps:.2f}, Effective Acceptance Rate: {accepted_steps/steps:.3f}")
    return n_sum / accepted_steps if accepted_steps > 0 else 0

beta_eps_values = np.linspace(0.1, 5.0, 20)
steps = 10000
original_results = []
modified_results = []
analytical_results = []

np.random.seed(42)
for beta_eps in beta_eps_values:
    n_orig = monte_carlo_n_original(beta_eps, steps)
    n_mod = monte_carlo_n_modified(beta_eps, steps)
    n_anal = analytical_n(beta_eps)
    original_results.append(n_orig)
    modified_results.append(n_mod)
    analytical_results.append(n_anal)

plt.figure(figsize=(10, 6))
plt.plot(beta_eps_values, analytical_results, 'b-', label='Analytical')
plt.plot(beta_eps_values, original_results, 'go', label='Original Scheme')
plt.plot(beta_eps_values, modified_results, 'ro', label='Modified Scheme')
plt.xlabel(r'$\beta \epsilon$')
plt.ylabel(r'$\langle n \rangle$')
plt.legend()
plt.grid(True)
plt.show()

print("\nÎ²Ïµ\tOriginal\tModified\tAnalytical\tMod. Rel. Error")
for b, orig, mod, an in zip(beta_eps_values, original_results, modified_results, analytical_results):
    rel_error = abs(mod - an) / an if an != 0 else 0
    print(f"{b:.2f}\t{orig:.4f}\t\t{mod:.4f}\t\t{an:.4f}\t\t{rel_error:.4f}")
```
