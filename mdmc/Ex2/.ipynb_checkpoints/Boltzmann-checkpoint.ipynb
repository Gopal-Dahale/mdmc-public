What do you note as the temperature is increased in terms of state occupancy? Does it make sense physically?

As the temperature increases, the occupancy distribution becomes more spread out, meaning higher-energy states are increasingly populated.
At low temperature, the system predominantly occupies the lowest energy state. At higher temperature, we have increased thermal energy, allowing particles to occupy higher-energy levels more frequently.

:::{admonition} Exercise 7
:class: exercise
Compare different degeneracies (no degeneracy, $s+1$, and $s+2$) for different reduced temperatures. What can you infer?
:::

**Low Temperature Regime ($T \approx 0.5$):** Degeneracy effects are minimal since the ground state is mostly occupied.

**Intermediate Temperature Regime ($T \approx 1,2$):** Systems with higher degeneracies ($s+1$ and $s+2$) show greater occupancy in higher energy states compared to the no-degeneracy case because degeneracy factors weight higher energy states more, making them more probable.

**High Temperature Regime ($T \approx 3$):**  The system approaches a nearly uniform distribution as $T \to \infty$, where all states become equally probable in the classical limit.

1. The more degenerate a state is, the more likely it is to be occupied, even if it has higher energy.
2. The system remains in the ground state regardless of degeneracy because thermal energy is too low to excite it.
3. The system transitions toward a classical-like behavior where the degeneracy dominates over the energy suppression.

```python
import sys

import matplotlib.pyplot as plt
import numpy as np

sys.path.append("..")
import helpers

helpers.set_style()
```

```python
# MODIFY HERE
# set number of energy levels and temperatures here
n_energy_levels = 10
reducedTemperatures = [0.5, 1, 2, 3]

# hbar * ω = 1
# En = hbar * ω (n + 1/2) => (n + 1/2)
# P(En) = (1/Z) * gn * exp(-beta * En) = (1/Z) * gn * exp(-beta * (n + 1/2))
# gn => degeneracy
# The partition function is calculated in the for loop so we only return
# the non-normalized quantity in `calculateStateOccupancy`.


def calculateStateOccupancy(T, i):
    # No degeneracy
    E = i + 0.5
    return np.exp(-E / T) if T > 0 else 0  # Avoid division by zero


def calculateStateOccupancy_s1(T, i):
    # Degeneracy s+1
    # Multiply by gi = (i + 1) to account for degeneracy
    return (i + 1) * calculateStateOccupancy(T, i)


def calculateStateOccupancy_s2(T, i):
    # Degeneracy s+2
    # Multiply by gi = (i + 2) to account for degeneracy
    return (i + 2) * calculateStateOccupancy(T, i)


functions = {
    "no_degeneracy": calculateStateOccupancy,
    "s+1": calculateStateOccupancy_s1,
    "s+2": calculateStateOccupancy_s2,
}

for f in functions.keys():
    fig, ax = plt.subplots(1)
    ax.set_title(f)
    ax.set_xlabel("Energy level")
    ax.set_ylabel("Occupancy")

    calculateOccupancy = functions[f]

    for reducedTemperature in reducedTemperatures:
        distribution = []  # For each state there is one entry
        partition_function = np.float64(0.0)

        for i in range(n_energy_levels):
            stateOccupancy = calculateOccupancy(reducedTemperature, i)
            distribution.append(stateOccupancy)  # MODIFY HERE
            partition_function += stateOccupancy  # MODIFY HERE

        ax.plot(distribution / partition_function, label=reducedTemperature)

    ax.legend()
    plt.savefig(f"{f}.png", bbox_inches="tight")
    plt.show()
```

```python
for f in functions.keys():
    fig, ax = plt.subplots(1)
    ax.set_title(f"{f} with {n_energy_levels} Energy Levels")
    ax.set_xlabel("Reduced Temperature")
    ax.set_ylabel("Partition Function")

    calculateOccupancy = functions[f]
    Z = []  # This variable will store the partition function
    # for different reduced temperatures

    for reducedTemperature in reducedTemperatures:
        distribution = []  # For each state there is one entry
        partition_function = np.float64(0.0)

        for i in range(n_energy_levels):
            stateOccupancy = calculateOccupancy(reducedTemperature, i)
            distribution.append(stateOccupancy)  # MODIFY HERE
            partition_function += stateOccupancy  # MODIFY HERE
        Z.append(partition_function)  # MODIFY HERE
    ax.plot(reducedTemperatures, Z, label="Z Exact")
    ax.plot(
        reducedTemperatures, 2 * reducedTemperatures, label="Z Approximated"
    )  # MODIFY HERE
    # insert approximated Z
    ax.legend()
    plt.savefig(f"{f}.png", bbox_inches="tight")
    plt.show()
```

```python
def calculateStateOccupancyVec(T, J):
    E = J * (J + 1) / 2
    E = E[:, np.newaxis]
    J = J[:, np.newaxis]
    return (2 * J + 1) * np.where(T > 0, np.exp(-E / T), 0)


fig, ax = plt.subplots(1, 2, figsize=(12, 5))
fig.suptitle("Linear Rotor")
ax[0].set_title("$Z_{{exact}}$")
ax[1].set_title("$|Z - Z_{{approx}}|$")
ax[0].set_xlabel("Reduced Temperature")
ax[0].set_ylabel("Partition Function")

ax[1].set_xlabel("Reduced Temperature")
ax[1].set_ylabel(r"$|Z - Z_{approx}|$")

Z = []  # This variable will store the partition function
# for different reduced temperatures

n_energy_levels = 50
reducedTemperatures = np.linspace(0.5, 100, 100)

stateOccupancy = calculateStateOccupancyVec(
    reducedTemperatures, np.arange(n_energy_levels)
)

# Initialize with zeros and same dtype
Z = np.zeros_like(stateOccupancy, dtype=stateOccupancy.dtype)

# Initialize row sum
cumulative_sum = np.zeros(len(reducedTemperatures), dtype=stateOccupancy.dtype)

for i in range(n_energy_levels):
    cumulative_sum += stateOccupancy[i]  # Accumulate the rows
    Z[i] = cumulative_sum.copy()  # Store the cumulative sum

for i, energy_level in enumerate(range(0, n_energy_levels, 10)):
    ax[0].plot(
        reducedTemperatures, Z[energy_level], label=f"{energy_level} Energy Levels"
    )
    ax[1].plot(
        reducedTemperatures,
        np.abs(Z[energy_level] - 2 * reducedTemperatures),
        label=f"{energy_level} Energy Levels",
    )

ax[0].plot(
    reducedTemperatures,
    2 * reducedTemperatures,
    color="black",
    linewidth=2.0,
    linestyle="--",
    label="Z Approximated",
)

# ax.set_yscale('log')
ax[0].legend()
ax[1].legend()
plt.savefig(f"Linear Rotor varying Energy Levels.png", bbox_inches="tight")
plt.show()
```
